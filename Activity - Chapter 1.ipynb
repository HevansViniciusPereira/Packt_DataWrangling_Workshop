# Activity 1

Create a list of 100 random numbers.

```python
import random
import numpy as np
```

```python
lista1 = [random.randint(0,100) for n in range(100)]
print(lista1)
```

Create a new list from this random list, with numbers that are divisible by 3.

```python
lista2 = [x for x in lista1 if x % 3 == 0]
print(lista2)
```

Calculate the length of these two lists and store the difference in a new variable.

```python
n1 = len(lista1)
n2 = len(lista2)

n = n1 - n2
n
```

Using a loop, perform steps 1, 2, and 3, and find the difference variable 10 times.

```python
length = []

for i in range(10):
    lista1 = [random.randint(0,100) for n in range(100)]
    lista2 = [x for x in lista1 if x % 3 == 0]
    n = len(lista1) - len(lista2)
    length.append(n)
    
length
```

Find the arithmetic mean of these 10 difference values.

```python
np.mean(length)
```

Find all of the words in multiline_text using the split method.

```python
words = multiline_text.split()
words
```

Create a list from this list that will contain only the unique words.

```python
list_1 = dict.fromkeys(words)
list_1
```

Count the number of times the unique word has appeared in the list using the key and value in dict.

```python
for word in words:
    if list_1[word] is None:
        list_1[word] = 1
    else:
        list_1[word] = list_1[word] + 1
        
list_1
```

Find the top 25 words from the unique words that you have found using the slice method.

```python
top25_words = sorted(list_1.items(), key=lambda key_val_tuple: key_val_tuple[1], reverse=True)
top25_words
```
